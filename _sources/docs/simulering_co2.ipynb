```python
import matplotlib.pyplot as plt 
import pandas as pd

tid_slutt = 2030  # År etter 1825
tid = 1825        # Startår
a = 0.053         # Utslippsrate
u = 0.01          # Utslipp i 1825 (tonn per innbygger)
b = 10.5

utslipp = [u]     # Legger inn startutslipper i en liste
årstall = [tid]   # Legger inn startåret i en liste

data = pd.read_csv("datafiler/CO2_Norge.txt")
t_data = data["år"]
u_data = data["co2"]

while tid <= tid_slutt: 
    u = u + a*u*(1 - u/b)
    tid = tid + 1
    if tid >= 2006:
        a = -0.1
    utslipp.append(u)
    årstall.append(tid)

plt.plot(årstall, utslipp, label = "Modell")
plt.plot(t_data, u_data, label = "Data")
plt.xlabel("Årstall")
plt.ylabel("Utslipp av CO2 i tonn/innbygger")
plt.legend()
plt.savefig("co2_ny.tiff")
```
